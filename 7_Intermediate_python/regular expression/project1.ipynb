Imagine you're building a website.People type emails like:

hanifa@gmail.com

wrong.@gmail..com

@gmail

hello@gmail

You need a smart gatekeeper to check: ‚ÄúIs this a VALID email format?‚Äù This gatekeeper is your RegEx pattern.

üí° WHAT are we trying to do?

We want to confirm:

‚úî email has username            ‚úî contains @
‚úî has a domain name             ‚úî ends with TLD (.com, .pk, .org, etc.)

üß† WHEN is this used?

Everywhere:

signup forms                  login pages               account recovery                 enterprise systems

user management systems       e-commerce                school management apps            all backend servers





```python
import re

email_pattern = re.compile(
    r"^[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[A-Za-z]{2,}$"
)

email = "hanifa@gmail.com"

if email_pattern.match(email):
    print("Valid Email ‚úî")
else:
    print("Invalid Email ‚ùå")
```
