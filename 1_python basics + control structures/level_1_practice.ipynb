```python
# 1: Ask the user for a number and tell if it's even or odd.
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")
```

```python
# 2: Input: x operator y (e.g., 10 * 3) --> Output result.
expression = input("Enter expression: ") 

# splitting expression
x, y, z = expression.split(" ")

# converting str into int
x = int(x)
z = int(z)

if y == "+":
    print(f"{expression} = ", x + z)
elif y =="-":
    print(f"{expression} = ", x - z)
elif y == "*":
    print(f"{expression} = ", x * z)
elif y == "/":
    print(f"{expression} = ", x / z)
elif y == "//":
    print(f"{expression} = ", x // z)
elif y == "%":
    print(f"{expression} = ", x % z)
elif y == "**":
    print(f"{expression} = ", x ** z)
else:
    print(f"{y}, Invalid operator! Use (+, -, *, /, //, %, **).")

```

```python
# Input: percent --> Output grade (A, B, C, fail) using if-elif-else.
marks = int(input("Enter  obtained percent: ")) 
total = int(input("Enter total percent: "))
percent = ((marks) / (total)) * 100
print("\n****  GRADE ****")
if percent >= 90:
 print(f" Obtained Marks = {marks} \n Total Marks = {total} \n  percentage {percent}% \n Grade: 'A'.")
elif percent >= 80:
 print(f" Obtained Marks = {marks} \n Total Marks = {total} \n  percentage {percent}% \n Grade: 'B'.")
elif percent >= 70:
 print(f" Obtained Marks = {marks} \n Total Marks = {total} \n  percentage {percent}% \n Grade: 'C'.")
elif percent >= 60:
 print(f" Obtained Marks = {marks} \n Total Marks = {total} \n  percentage {percent}% \n Grade: 'D'.")
elif percent >= 50:
 print(f" Obtained Marks = {marks} \n Total Marks = {total} \n  percentage {percent}% \n Grade: 'E'.")
else: 
 print(f" Obtained Marks = {marks}  \n Total Marks = {total} \n  percentage {percent}% \n Grade: 'fail'.")
```

```python
# Grading system using ' match case '
marks = int(input("Enter  obtained percent: ")) 
total = int(input("Enter total marks: "))
p = ((marks) / (total)) * 100
print("\n****  GRADE ****")
match p:
    case p if p >= 90:
        print("Grade: A")
    case p if p >= 80:
        print("Grade: B")
    case p if p >= 70:
        print("Grade: C")
    case p if p >= 60:
        print("Grade: D")
    case p if p >= 50:
        print("Grade: E")
    case _:
        print("Grade: Fail")


```
