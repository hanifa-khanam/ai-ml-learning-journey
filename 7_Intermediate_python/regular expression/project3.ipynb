```python
import re

def is_strong_password(password):
    pattern = re.compile(
        r"""^(?=.*[A-Z])       # one uppercase
            (?=.*[a-z])        # one lowercase
            (?=.*\d)           # one digit
            (?=.*[@$!%*?&])    # one special char
            [A-Za-z\d@$!%*?&]{8,}$   # allowed + length
        """, 
        re.VERBOSE
    )
    
    return bool(pattern.match(password))


# Test
tests = ["Hanifa123!", "weakpass", "HELLO123", "hanifa@", "Strong@2025"]
for t in tests:
    print(t, "=>", is_strong_password(t))
```
