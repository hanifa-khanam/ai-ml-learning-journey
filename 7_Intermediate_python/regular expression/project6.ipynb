Imagine you are analyzing a document, chat, or database full of mixed text.
Your job is to find all the Pakistani phone numbers hidden inside.

The format you want is:

03XX-XXXXXXX


For example:

0301-2345678

0335-9876543

0342-0011223

Your RegEx robot scans the text and extracts all valid phone numbers.


**ðŸ§  Why?**

Used in:

WhatsApp scrapers

telecom applications

cleaning messy datasets

```python
import re

phone_pattern = re.compile(r"03\d{2}-\d{7}")

def extract_phone_numbers(text):
    return phone_pattern.findall(text)


# Test
text = """
Contact us at 0301-2345678 or 0335-9876543.
Invalid: 301-2345678, 03012345678, 0342-123456.
"""

print(extract_phone_numbers(text))
```
