```python
# Multiplication Table Generator
# Input: number --> print multiplication table up to 10.
n = int(input("Enter number: "))
for i in range(0, 11):
    table = n * i
    # i += 1
    print(f"{n} x {i} =  ", table)
```

```python
# Sum of digits
# Input: 1234 --> Output: 1 + 2 + 3 + 4 = 10
user = input("Enter number: ")
total = 0
for digit in user:
    total += int(digit)
print(f"Sum of digits {user} = ", total)
    
```

```python
# Prime Number Checker
# Check if a number is prime using loops.
n = int(input("Enter a number: "))
if n <= 1:
    print(f"{n} is not a Prime Number.")
else:  
    for i in range(2, int(n**0.5)+1):
        if n % i == 0 :
            print(f"{n} is not a Prime Number.")
            break
    else:
        print(f"{n} is a Prime Number.")
    

        
```

```python
# Guess the number Game 
# Program picks a random number (1-20). User keeps guessing until correct.
import random
r = random.randint(1, 21)
while True:
    n = int(input("Enter a number: "))
    if n == r: 
        print("You Win!\nCorrect number: ", r)
        break
    else:
        print("You Guess Incorrect number! Try again.")


```
