```python
# ATM Simulator
""" Ask for PIN (with 3 chances).
    Show menu: check balance, deposit, withdraw.
    Use loops + conditionals. """
# pin = '123456'
# while True:
#     user = input("Enter your PIN: ")
#     if user == pin:
#         print("****Successfully login.****")
#         break
#     else:
#         print("Invalid PIN! Try again.")
pin = "123456"
attempts = 3
# -----PIN check with 3 chances -----
while attempts > 0:
    user = input("Enter your PIN: ")
    if user == pin:
        print("**** Successfully Login ****")
        break
    else:
        attempts -=1
        print(f"Invalid PIN! You have {attempts} attempts left.")
else:
    print(f"Too many wrong attempts! Exiting....")
    exit()

balance = 15000
if user == pin:
    print("## MENU ##")
    print("What do you want to check?")
    print("1: Check balance.\n2: Deposit.\n3: Withdraw.\n4: Exit.")
    print("Choose option (1/2/3/4).")
    choice = input("Enter your choice (1/2/3/4): ")
    print(f"Your choice is option {choice}.")
    if choice == '1':
        print(f"Balance:\n Your current Balance = ${balance}")
    elif choice == '2':
        deposit = int(input("Enter your deposit amount: "))
        balance1 = balance + deposit
        print(f"Deposit:\n Your amount before deposit = ${balance}\n Your amount after deposit = ${balance1}")
    elif choice == '3':
        withdraw = int(input("Enter your withdrawal amount: "))
        if withdraw > balance:
            print(f"Insufficient balance! withdrawal failed.")
        else:
            balance2 = balance - withdraw
            print(f"Withdraw:\n Your amount before Withdrawal = ${balance}\n Your amount after Withdrawal = ${balance2}")
    elif choice == '4':
        print("Exiting! Good Bye.")
    else:
        print(f"Invalid Choice! choose (1/2/3/4).")


```
