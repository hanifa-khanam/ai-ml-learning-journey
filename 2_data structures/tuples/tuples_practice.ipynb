**P6: Convert a list of fruits into a tuple, and check if "apple" exits.**

```python
fruits_lst = ['apple', 'banana', 'mango', 'orange']

fruits_tpl = tuple(fruits_lst)
print(fruits_tpl)

if 'apple' in fruits_tpl:
    print(f"Yes 'apple' is in the list.")
else:
    print(f"'apple' is not in the list.")
```

## **Level 3: Final Mini Project**

**Write a program:**

**. Ask the user to input 3 coordinates (x, y, z)**

**. Store each as a tuple inside a list.**

**. Print all coordinates and coculate how many are at the origin (0, 0, 0)**

```python
# Empty list to store coordinates
lst = []

limit = 3   # number of coordinates to take
while limit > 0:
    user_input = input("Enter 3 coordinates (x, y, z) separated by commas: ")

    # convert input into a tuple of integers
    coords = tuple(map(int, user_input.split(",")))
    
    lst.append(coords)  # add tuple into list
    limit -= 1

# Print all coordinates
print("List of coordinates: ", lst)

# Count how many are at the origin (0,0,0)
origin_count = lst.count((0, 0, 0))
print(f"Number of coordinates at the origin (0,0,0): {origin_count}")
```
