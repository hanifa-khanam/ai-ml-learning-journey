üåü Story ‚Äî The Web Detective

Imagine you are building a web scraper or AI model that reads text from the internet.
Inside the text, you want to find all URLs ‚Äî links to websites, blogs, images, pages, etc.

A special web detective walks through the text and picks out every URL he finds.

This detective is your URL Extractor powered by RegEx.

**üß† Why?**

Useful for:

AI dataset cleaning

content extraction

SEO analysis

detecting spam links

data mining

building crawlers

```python
import re

url_pattern = re.compile(
    r"(https?://[^\s]+|www\.[^\s]+)"
)

def extract_urls(text):
    return url_pattern.findall(text)
0

# Test
text = """
Visit https://vu.edu.pk for admissions.
Also check www.google.com or http://example.org/page.
"""

print(extract_urls(text))
```
