## **level 1: Basics of lists**

**P1: Ask the user to enter 5 favorite colors, store them in a list, then print them one by one.**

```python
# empty list
fav_color = []
# taking input from user
user = input("Enter your 5 favorite  colors (separeted by commas): ").split(",")

fav_color = user

fav_color = [color.strip() for color in fav_color]

# printing list
print("User favorite color list: ", fav_color)

# iterating through list to print each color
for color in fav_color:
    print(color)
```

**P2: Take a list of numbers [1, 2, 3, 4, 5] and print their squares using a loop.**

```python
num = [1, 2, 3, 4, 5]

# using list comprehension
num2 = [n**2 for n in num]
print("list comprehension: ", num2)

# using loops 
num2_loop = []
for n in num:
    num2_loop.append(n**2)
print("Loop result: ", num2_loop)
```

## **level 2: Adding and removing**

**P3: Keep asking user for guest names and add them to a list until they type "stop". Finally, print the guest list.**

```python
guest_lst = []
limit = 'stop'
 
while True:
    
    user1 = input("Enter guest names: (separated by commas): ")

    
    if user1.lower() == limit:
        break
    else:
        guest_lst.extend([name.strip() for name in user1.split(",")])

print("Guset list: ", guest_lst)
    

```

**P4: Start with a list of fruits ["apple", "banana", "cherry"].**

**Ask the user which fruit to remove, and delete it if it exists.**

```python
fruits = ['apple', 'banana', 'cherry']
print("Fruits: ", fruits)

print("Which fruit you want to remove from the fruit list?")
user2 = input("Your answer: ").lower()
print("Your answer: ", user2)

if user2 in fruits:
    fruits.remove(user2)
    print("Updated Fruit list: ", fruits)
else:
    print(f"{user2} is not found in the fruits list.")
```

## **level 3: Searching and Counting**

**P5: Given a sentence, split it into words and count how many times "the" appears.**

**P6: Create a list of numbers. Ask the user for a number and tell them the index if it exists.**

```python
numbers = [3, 6, 9, 34, 656, 87, 23, 67]
print("List of numbers: ", numbers)

print("What number you want to search?")
user4 =  int(input("Your answer: "))
print("Your answer: ", user4)

if user4 in numbers:
    print(f"Your number {user4} is found at index: ", numbers.index(user4))

else:
    print(f"{user4} is not found in the list.")
```

## **level 4: Sorting and Copying**

**P7: Input student marks into a list, then show them in ascending and descending order.**

```python
marks = []
until = 'stop'

while True:
    student = (input("Enter student marks: "))

    if student == until:
        break
    else:
        marks.append(student)

print(f"Student marks list: ", marks)
marks.sort()
print(f"Student marks list from top: ", marks)
```

**P8: Create a list of 5 items, make a copy, modify the copy, and show that the original list didn't change.**

```python
items = ['apple', 78, 'juice', 34, 56]

b = items.copy()
print("original: ", items)
print("copy:",b)
b.append("ah")
print("original: ", items)
print("copy: ", b)
b.remove('apple')
print("original: ", items)
print("copy: " ,b)
b.pop()
print("original: ", items)
print("copy: " ,b)
```

**P9: Make a list of lists (2D list), like [[1,2,3],[4,5,6],[7,8,9]].**

**Print it row by row in matrix form.**

```python
list_2d = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# print("First row: ", list_2d[0])
# print("Second row: ", list_2d[1])
# print("Third row: ", list_2d[2])

print("Matrix Form: ")
for row in list_2d:
    print(row)
print()

# without brackets
print("Matrix form(clean): ")
for row in list_2d:
    for item in row:
        print(item, end=' ')
    print()
```
