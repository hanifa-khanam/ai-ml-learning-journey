## **level 1: Basics and Indexing**

*P1: Write a program to take a string input from the user and print:*

*. The first Character*

*. The last Character*

*. The String in reverse*


```python
user = input("Enter a word: ")
print("Input of user : ", user)
# printing first character
first_char = user[0]   # index 0 = first character
print(f"First character :  ", first_char)

# printing last character
last_char = user[-1]    # index -1 = last character
print(f"Last character :  ", last_char)

# printing reverse of string
str_reverse = user[::-1]  # revers the string = ::-1
print(f"Reverse :  ", str_reverse)

```

P3: Take a string and replace the middle character with *. 

```python
str_ing = input("Enter a word: ")
print("Input of user: ", str_ing)

# find middle index
mid = len(str_ing) // 2

# replace middle character with *
rep_str = str_ing[:mid] + "*" + str_ing[mid+1:]
print(f"String after replacement: ", rep_str)
```

## **level 2: Case Handling**


 P4: Input a sentence and print it in:

```python
sentence = input("Enter sentence: ")
print(f"User Input sentence: ", sentence)

# uppercase
sen1 = sentence.upper()
print(f"Upper case : ", sen1)

# lowercase
sen2 = sentence.lower()
print(f"Lower case: ", sen2)

# title case
sen3 = sentence.title()
print(f"Title case: ", sen3)
```

P24: Write a program that counts vowels and consonants separately.

```python
def main():
    enter = input("Enter a sentence: ").lower()
    print("Input: ", enter)
    vowel_counter(enter)
    consonant_counter(enter)

def vowel_counter(enter):
    vowel = "aeiou"
    count = 0
    for char in enter:
        if char in vowel:
            count += 1
    print(f"vowels = ", count)       
        
def consonant_counter(enter):
    consonant = "bcdfghjklmnpqrstvwxyz"
    total = 0
    for char in enter:
        if char in consonant:
            total += 1
    print("Consonant: ", total)

main()

```

P25: Check if a given string is a palindrome (same forward and backword)

```python
given = input("Enter a word: ").lower()
print("Input string: ", given)

palindrome = given[::-1]
if given == palindrome:
    print(f"Yes, the word '{given}' is Palindrome.")
else:
    print(f"No, the given word '{given}' is not a Palindrome.")

    
    
```
