**P2: Given a = {1, 2, 3} and b = {2, 3, 4}, compute:**

**union, intersection, difference, symmetric difference.**

```python
a = {1, 2, 3}
b = {2, 3, 4}

# union
print(f"Union of a = {a} and b = {b}: \n ", a.union(b))

# intersection
print(f"Intersection of a = {a} and b = {b}: \n ", a.intersection(b))

# difference
print(f"Difference of a = {a} from b = {b}: \n ", a.difference(b))
print(f"Difference of b = {b} from a = {a}: \n ", b.difference(a))

# symmetric difference
print(f"Symmetric difference between a = {a} and b = {b}: \n ", a.symmetric_difference(b))
```

**P3: Write a code to check whether a list L contains duplicates (print True/False).**

```python
L  = ['a', 'b', 'a', 'c']

S = set(L)

if L == S:
    print("False, no duplicates.")
else:
    print("True, duplicates found. ")
```

**P4: Given two attendee lists for two envents, print the people who went to both events.**

```python
event_1 = {'Ali', 'Amna', 'Hamza', 'Nouman', 'Raheela', 'Haris'}
event_2 = {'Nouman', 'Haris', 'Raheela', 'Hanifa'}

# people who attend both events
print(f"Attend both events: ", event_1.intersection(event_2))
```

### **Mini Project**

**Unique words and common word**

> Ask user for two sentences.

> Produce:

set of unique words in sentence 1

set of unique words in sentence 2

words appearing in both(intersection)

words only in sentence 1(difference)



```python
sntnce1 = input("Enter your first sentence: ").casefold()
sntnce2 = input("Enter your second sentence: ").casefold()

s1 = set(sntnce1.split())
s2 = set(sntnce2.split())

print("First Sentence: ", sntnce1)
print("Set from 1st sentence: ", s1)
print()
print("Second Sentence: ", sntnce2)
print("Set from 2nd sentence: ", s2)

print()

# words appearing in both 
print("Common words: ", s1 & s2)

print()

# word only in sentence 1
print("words only in sentence 1: ", s1 - s2)



```
