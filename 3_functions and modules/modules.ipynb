## math module

```python
import math

print(math.sqrt(25))      # Square root
print(math.pow(2, 3))     # Power
print(math.pi)            # π constant
print(math.sin(math.radians(90)))  # Sine of 90 degrees
```

```python
import math

angle = 90
radians = math.radians(angle)  # Convert degree → radian
print("Sine:", math.sin(radians))
print("Cosine:", math.cos(radians))
print("Tangent:", math.tan(radians))
```

```python
import os
import shutil

# Ask user for path or file/folder name
path = input("Enter file or folder name to delete: ").strip()

# Check if it exists first
if os.path.exists(path):
    try:
        # If it's a file
        if os.path.isfile(path):
            os.remove(path)
            print(f" File '{path}' deleted successfully!")

        # If it's a folder
        elif os.path.isdir(path):
            shutil.rmtree(path)
            print(f"Folder '{path}' and its contents deleted successfully!")

    except PermissionError:
        print(f"Access denied: Close any open files/folders related to '{path}' and try again.")
    except Exception as e:
        print(f"Unexpected error: {e}")

else:
    print(f"'{path}' does not exist.")
```

```python
import os
import shutil

path = input("Enter the file or folder name to diagnose: ").strip()

# 1Check if the path exists
if not os.path.exists(path):
    print(f"'{path}' does not exist in current directory.")
else:
    print(f"Found: '{path}'")

    #  Print current working directory
    cwd = os.getcwd()
    print(f"Current working directory: {cwd}")

    #  Show absolute path of target
    abs_path = os.path.abspath(path)
    print(f"Target absolute path: {abs_path}")

    #  Check if your current directory is inside the target folder
    if cwd.startswith(abs_path):
        print("Your current working directory is INSIDE the folder you are trying to delete.")
        print("Move one directory up before deleting.")
    else:
        print("You are NOT inside the target folder.")

    #  Check if it’s a file or folder
    if os.path.isfile(path):
        print("Type: File")
    elif os.path.isdir(path):
        print("Type: Folder")

        #  List its contents (hidden files can cause permission issues)
        contents = os.listdir(path)
        if contents:
            print(f"Folder contains {len(contents)} items:")
            for item in contents:
                print(f"  - {item}")
        else:
            print("Folder is empty.")
    
    #  Optional: ask if user wants to delete
    choice = input("\nDo you want to delete it now? (Y/N): ").lower()
    if choice == "y":
        try:
            if os.path.isfile(path):
                os.remove(path)
                print(f"File '{path}' deleted successfully.")
            elif os.path.isdir(path):
                # Move up to avoid locking
                os.chdir("..")
                shutil.rmtree(path)
                print(f"Folder '{path}' deleted successfully.")
        except PermissionError:
            print(f"Access denied. Close files or VS Code windows that use '{path}' and try again.")
        except Exception as e:
            print(f"Unexpected error: {e}")
    else:
        print("Deletion cancelled.")
```

## sys module


```python

import sys

sys.argv = ['temp_converter.py', 'C', '37']  # Simulate terminal input

def c_to_f(c):
    return (c * 9/5) + 32

def f_to_c(f):
    return (f - 32) * 5/9

if len(sys.argv) != 3:
    sys.exit(" Usage: python temp_converter.py <C/F> <value>")

unit = sys.argv[1].upper()
value = float(sys.argv[2])

if unit == "C":
    print(f"{value}°C = {c_to_f(value):.2f}°F")
elif unit == "F":
    print(f"{value}°F = {f_to_c(value):.2f}°C")
else:
    sys.exit(" Please enter unit as 'C' or 'F'.")
```
