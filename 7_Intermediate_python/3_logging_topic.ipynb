## **Story & Analogy**
Imagine managing a huge hotel. Every action â€” guest check-in, room cleaning, payments â€” must be tracked. You hire a **recorder** to log events. Later, if something goes wrong, you check the logs to debug issues.  
In Python, the **logging module** is that recorder. It keeps track of program events professionally, unlike `print()`.

---

## **What is Logging?**
Logging is tracking events that happen during program execution. It helps in:
- Debugging
- Monitoring program flow
- Recording errors and exceptions
- Maintaining permanent logs in files

---

## **Why Logging?**
- Debug complex programs efficiently
- Provides audit trails
- Avoids cluttered `print()` statements
- Allows filtering by severity levels

---

## **Logging Levels**

| Level     | Description                         | When to Use                           |
|----------|-------------------------------------|---------------------------------------|
| DEBUG    | Detailed info for diagnosing issues | During development/testing            |
| INFO     | General program info                 | Track normal flow                      |
| WARNING  | Unexpected situation, not critical  | Potential issues                        |
| ERROR    | Serious problem occurred             | Exceptions, failed operations          |
| CRITICAL | Very severe errors                   | System crash or critical exceptions    |

---



## **Basic Logging Setup**

```python
import logging

logging.basicConfig(level=logging.INFO)
logging.debug("Debug message")       # Ignored, lower than INFO
logging.info("Program started")      # Displayed
logging.warning("Warning!")          # Displayed
logging.error("An error occurred")   # Displayed
logging.critical("Critical problem") # Displayed


# **Logging to a File**

````python
import logging

logging.basicConfig(
    filename='app.log',
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

logging.info("Program started")
logging.error("An error occurred")
````

# **Advanced: Custom Logger with Handlers**

````python
import logging

logger = logging.getLogger("HotelLogger")
logger.setLevel(logging.DEBUG)

# File handler (logs WARNING and above)
fh = logging.FileHandler("hotel.log")
fh.setLevel(logging.WARNING)

# Console handler (logs INFO and above)
ch = logging.StreamHandler()
ch.setLevel(logging.INFO)

# Formatter
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
fh.setFormatter(formatter)
ch.setFormatter(formatter)

# Add handlers
logger.addHandler(fh)
logger.addHandler(ch)

# Logging examples
logger.debug("Debug message")   # Console only
logger.info("Info message")     # Console only
logger.warning("Warning!")      # Console + file
logger.error("Error!")          # Console + file
````

# **Exception Logging**

Logs stack trace automatically

Use inside try-except blocks

````python
try:
    result = 10 / 0
except Exception as e:
    logging.exception("An error occurred")
````

# **Rotating & Timed Logs**

Rotating by Size

````python
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler('hotel.log', maxBytes=1024, backupCount=3)
````

Rotating by Time

````python
from logging.handlers import TimedRotatingFileHandler

handler = TimedRotatingFileHandler('hotel_timed.log', when='D', interval=1, backupCount=7)
````

# **Filtering Logs**

By Level

````python
class WarningFilter(logging.Filter):
    def filter(self, record):
        return record.levelno >= logging.WARNING
````

By Module / Custom

````python
class ModuleFilter(logging.Filter):
    def __init__(self, module_name):
        self.module_name = module_name
    def filter(self, record):
        return record.name == self.module_name
````
