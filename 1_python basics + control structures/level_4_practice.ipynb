```python
# Vowel Counter 
# count how many vowels are in a given string.
user = input("Enter a word: ").lower()
count = 0
vowel = ('a', 'e', 'i', 'o', 'u')
for char in user:
    if char in vowel:
        count += 1
print(f"There are {count} vowels in your word '{user}'.")

```

```python
# Binary calculator
# Take two numbers and perform bitwise operations (&, |, ^, ~).
a = int(input("Enter a : "))
b = int(input("Enter b : "))
# bitwise and (&) operation
print(f"Bitwise AND (&) : {a} & {b} = ", a & b, "= " , bin(a & b))
# bitwise or (|) operation
print(f"Bitwise OR (|) : {a} | {b} =  ", a | b,  "= " , bin(a | b))
# bitwise xor (^) operation
print(f"Bitwise XOR (^) : {a} ^ {b} = ", a ^ b,  "= " , bin(a ^ b)) 
# bitwise not (~) operation
print(f"Bitwise NOT (~) : ~{a} = ", ~a,  "= " , bin(~a))
print(f"Bitwise NOT (~) : ~{b} = ", ~b,  "= " , bin(~b))

```
