Imagine there is a digital gatekeeper whose job is to check if a CNIC number is written correctly.
A valid CNIC is always:

```python
import re

cnic_pattern = re.compile(
    r"^\d{5}-\d{7}-\d$"
)

def is_valid_cnic(cnic):
    return bool(cnic_pattern.match(cnic))


# Test
tests = [
    "31203-0580123-7",
    "45102-1234567-1",
    "1234-1234567-1",
    "31203-580123-7",
    "Hello-World",
]

for t in tests:
    print(t, "=>", is_valid_cnic(t))
```
